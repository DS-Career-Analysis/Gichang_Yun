```python
import pandas as pd
import re
import json
import pandas as pd
import requests
from tqdm import tqdm
import numpy as np
import folium
from folium import plugins

coordinate = pd.read_csv('coordinate.csv', encoding='utf-8')
coordinate.head()
```

```python
m = folium.Map(location=[35.9078, 127.7669], zoom_start=6, tiles='openStreetMap')

for i in range(len(coordinate)):
    folium.Marker(location=[coordinate['y'][i], coordinate['x'][i]], popup=coordinate['회사명'][i], tooltip=coordinate['road_address'][i]).add_to(m)

m
```
